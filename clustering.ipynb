```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Load image
image_path = '/content/OIP.jpg'  # replace with your image path
image = Image.open(image_path)
image = np.array(image, dtype=np.float64) / 255

# Reshape the image to be a list of RGB values
w, h, d = tuple(image.shape)
image_array = np.reshape(image, (w * h, d))

# Perform k-means clustering to form k clusters
k =1 # replace with your desired number of clusters
kmeans = KMeans(n_clusters=k).fit(image_array)

# Replace each pixel with its nearby centroid
labels = kmeans.predict(image_array)
new_image = kmeans.cluster_centers_[labels]
new_image = np.reshape(new_image, (w, h, d))

# Display the clustered image
plt.imshow(new_image)
plt.show()
```
